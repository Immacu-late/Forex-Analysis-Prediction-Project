#### Transactions Per Day

```python
# Extract day of the week from timestamp
df['day_of_week'] = df['timestamp'].dt.dayofweek

# Group by day of the week and count transactions
transactions_by_day = df.groupby('day_of_week').size()

# Calculate percentages
total_transactions = transactions_by_day.sum()
percentages = (transactions_by_day / total_transactions) * 100

# Plot bar chart with percentages
plt.figure(figsize=(8, 6))
ax = transactions_by_day.plot(kind='bar', color='skyblue')
plt.title('Transactions by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Transaction Count')
plt.xticks(range(7), ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.ylim(0, max(transactions_by_day) * 1.1)  # Adjust y-axis limit to accommodate percentages

# Highlight Thursday in a different color
highlight_color = 'orange'
thursday_index = 3  # Thursday index
for i, v in enumerate(percentages):
    color = highlight_color if i == thursday_index else 'skyblue'
    ax.patches[i].set_facecolor(color)  # Set color for each bar
    ax.text(i, transactions_by_day[i] + 0.5, f'{v:.1f}%', ha='center', va='bottom', color='black')  # Display percentage above each bar

plt.show()
```

From the bar graph above, we note that Thursday has the highest number of transactions in a week

#### Transactions Per Hour

```python
def plot_transactions_by_hour(df):
    """
    Plot transactions by hour of the day.
    
    Args:
    df (DataFrame): DataFrame containing transaction data with a 'timestamp' column.
    """
    # Extract hour from timestamp
    df['hour'] = df['timestamp'].dt.hour

    # Group by hour and count transactions
    transactions_by_hour = df.groupby('hour').size()

    # Plot bar chart
    plt.figure(figsize=(10, 6))
    transactions_by_hour.plot(kind='bar', color='orange')
    plt.title('Transactions by Hour of the Day')
    plt.xlabel('Hour of the Day')
    plt.ylabel('Transaction Count')
    plt.xticks(range(24))
    plt.show()
```

```python
plot_transactions_by_hour(df)
```

From the bar graph above, we note that highest number of transactions is between 9am to 6pm and lowest between 1am to 3am

#### Transactions Per Month

```python
# Extract month from timestamp
df['month'] = df['timestamp'].dt.month_name()

# Group by month and calculate total transaction amount
transactions_by_month = df.groupby('month')['destination_amount_as_usd'].sum()

# Calculate total transaction amount for all months
total_transactions = transactions_by_month.sum()

# Calculate percentage for each month
percentage_by_month = (transactions_by_month / total_transactions) * 100

# Plot bar chart
plt.figure(figsize=(10, 6))
bar_plot = transactions_by_month.plot(kind='bar', color='purple')
plt.title('Total Transaction Amount by Month')
plt.xlabel('Month')
plt.ylabel('Total Transaction Amount (USD)')

# Add percentage labels to the bars
for i, value in enumerate(transactions_by_month):
    plt.text(i, value, f'{percentage_by_month.iloc[i]:.2f}%', ha='center', va='bottom')

plt.show()
```

**Finding outliers using IQR Method**

Since the data doesn’t follow a normal distribution, we will calculate the outlier data points using the statistical method called interquartile range (IQR).

Using the IQR, the outlier data points are the ones falling below Q1–1.5 IQR or above Q3 + 1.5 IQR.
The Q1 is the 25th percentile and Q3 is the 75th percentile of the dataset, and IQR represents the interquartile range calculated by Q3 minus Q1 (Q3–Q1). 

```python
# Step 1: Filter out non-numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Step 2: Calculate the IQR for each numerical column
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

# Step 3: Determine lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 4: Identify outliers
outliers = ((numeric_df < lower_bound) | (numeric_df > upper_bound)).any(axis=1)

# Display outliers
print("Rows with outliers:")
print(df[outliers])
```

### Define X and Y variables

### Splitting Data

```python
#Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state = 42)

#Define the scaler
scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

### Model Building Evaluation

```python
def train_and_evaluate_model(model):
    #Fit the model
    model.fit(X_train_scaled, y_train)
    
    #Make a prediction
    y_pred = model.predict(X_test_scaled)
    
    #Evaluate the model
    MAE = mean_absolute_error(y_test, y_pred)
    MSE = mean_squared_error(y_test, y_pred)
    RMSE = np.sqrt(mean_squared_error(y_test, y_pred))
    r_squared = r2_score(y_test, y_pred)
    
    return model, MAE, MSE, RMSE, r_squared
```

```python
linear_model, lr_MAE, lr_MSE, lr_RMSE, lr_squared = train_and_evaluate_model(LinearRegression())
print(f'Linear Regression - MAE :{lr_MAE}, MSE : {lr_MSE}, RMSE : {lr_RMSE}, R2 : {lr_squared}')
```

```python
dt_model, dt_MAE, dt_MSE, dt_RMSE, dt_squared = train_and_evaluate_model(DecisionTreeRegressor(random_state = 42, max_depth = 4))
print(f'Decision Tree - MAE :{dt_MAE}, MSE : {dt_MSE}, RMSE : {dt_RMSE}, R2 : {dt_squared}')
```

```python
rf_model, rf_MAE, rf_MSE, rf_RMSE, rf_squared = train_and_evaluate_model(RandomForestRegressor(random_state=42,max_depth =4))
print(f'Random Forest - MAE :{rf_MAE}, MSE : {rf_MSE}, RMSE : {rf_RMSE}, R2 : {rf_squared}')
```

```python
lor_model, lor_MAE, lor_MSE, lor_RMSE, lor_squared = train_and_evaluate_model(LogisticRegression())
print(f'Logistic Regression - MAE :{lor_MAE}, MSE : {lor_MSE}, RMSE : {lor_RMSE}, R2 : {lor_squared}')
```

### Hyperparameters_ Linear Regression

```python
# Set up the data
X = encoded_df.drop('gross_profit_as_usd', axis=1)
y = encoded_df['gross_profit_as_usd']

# StandardScaler and LinearRegression
std_slc = StandardScaler()
lr = LinearRegression()

# Pipeline with StandardScaler and LinearRegression
pipe = Pipeline(steps=[('std_slc', std_slc), ('lr', lr)])

# Define parameters for grid search
param_grid = {
    'lr__fit_intercept': [True, False],
    'lr__normalize': [True, False]
}

# Initialize GridSearchCV
clf_GS = GridSearchCV(estimator=pipe, param_grid=param_grid, cv=5, verbose=1, n_jobs=-1)

# Fit GridSearchCV
clf_GS.fit(X, y)

# Print best parameters and best score
print("Best parameters found: ", clf_GS.best_params_)
print("Best score: ", clf_GS.best_score_)
```

```python
# Measure the time taken to fit the model
start_time = time.time()
clf_GS.fit(X, y)
end_time = time.time()
run_time = end_time - start_time

print("Time taken to fit the model: {:.2f} seconds".format(run_time))
```
